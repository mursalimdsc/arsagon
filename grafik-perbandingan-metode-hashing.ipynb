```python
import matplotlib.pyplot as plt
import numpy as np

# --- Data rata-rata waktu login dari tabel Anda (dalam ms) ---
average_login_times = {
    'Arsagon': 186.49,
    'Bcrypt': 308.01,
    'PBKDF2': 263.67,
    'Scrypt': 206.42
}

# --- Konfigurasi Grafik ---
methods = list(average_login_times.keys())
times = list(average_login_times.values())

# --- Membuat Grafik Batang ---
fig, ax = plt.subplots(figsize=(10, 7))

# Membuat batang
bars = ax.bar(methods, times, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])

# --- Menambahkan Label, Judul, dan Estetika ---
ax.set_ylabel('Average Login Time (ms)', fontsize=14)
ax.set_title('Average Login Time Comparison by Hashing Method', fontsize=16, pad=20)

# Mengatur sumbu X dan ukuran font labelnya
ax.tick_params(axis='x', labelsize=14) # Mengatur ukuran font label sumbu X

# Menambahkan label data di atas setiap batang
ax.bar_label(bars, fmt='%.2f', padding=3)

# Menambahkan grid untuk kemudahan membaca
ax.yaxis.grid(True, linestyle='--', alpha=0.7)
ax.set_axisbelow(True)

fig.tight_layout()

# --- Menyimpan Hasil Grafik ---
output_filename = 'average_login_time_comparison.png'
plt.savefig(output_filename, dpi=300)

print(f"Grafik telah berhasil disimpan sebagai '{output_filename}'")

# --- Menampilkan Grafik ---
plt.show()
```
